```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
folder = 'data/MachinedPart02'
img_part = folder+'/part02.jpg'
```

```python
img = cv2.imread(img_part,cv2.IMREAD_COLOR)
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
ok_imgs = os.listdir(folder+'/Ok')
defects_imgs = os.listdir(folder+'/Defects')
```

```python
fig, axs = plt.subplots(2,3, figsize=(15,8))

for i in range(0,3):
    img = plt.imread(folder+'/Ok/'+ok_imgs[i])
    axs[0][i].imshow(img)
    axs[0][i].axis('off')   
#axs[0][0].set_title('Ok')

for i in range(0,3):
    img = plt.imread(folder+'/Defects/'+defects_imgs[i])
    axs[1][i].imshow(img)
    axs[1][i].axis('off')   
#axs[0][1].set_title('Defects')
    
fig.tight_layout()
plt.savefig('img_dataset.png', dpi=200, bbox_inches='tight', pad_inches=0)
plt.show()
```
