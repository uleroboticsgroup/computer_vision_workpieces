```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import os
from skimage.feature.texture import greycoprops, greycomatrix
import numpy as np
import seaborn as sns
import math
```

```python
folder = 'data/MachinedPart02'
img_part = folder+'/part02.jpg'
```

```python
ok_imgs = os.listdir(folder+'/Ok')
defects_imgs = os.listdir(folder+'/Defects')
```

```python
props = ['contrast', 'dissimilarity', 'correlation', 'homogeneity', 'energy', 'ASM', ]
props_reduced = ['cont', 'diss', 'hom', 'ene', 'ASM', 'corr']
radians = np.concatenate((np.arange(0, np.pi, np.pi/8), [np.pi]))
degrees = [math.degrees(value) for value in radians] 
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8, 8))
```

```python
img_ok = plt.imread(folder+'/Ok/'+ok_imgs[0])
gray_ok = cv2.cvtColor(img_ok, cv2.COLOR_BGR2GRAY)
clahe_ok = clahe.apply(gray_ok)

g = greycomatrix(gray_ok, [3], radians) #distances, angles
ok_properties = {}
for prop in props:
    propperty = greycoprops(g, prop)
    ok_properties[prop] = propperty[0,:]
    
g = greycomatrix(clahe_ok, [3], radians) #distances, angles
for prop in props:
    propperty = greycoprops(g, prop)
    ok_properties[prop+'_clahe'] = propperty[0,:]
```

```python
img_defect = plt.imread(folder+'/Defects/'+defects_imgs[0])
gray_defect = cv2.cvtColor(img_defect, cv2.COLOR_BGR2GRAY)
clahe_defect = clahe.apply(gray_defect)

g = greycomatrix(gray_defect, [3], radians) #distances, angles
defect_properties = {}
for prop in props:
    propperty = greycoprops(g, prop)
    defect_properties[prop] = propperty[0,:]
    
g = greycomatrix(clahe_defect, [3], radians) #distances, angles
for prop in props:
    propperty = greycoprops(g, prop)
    defect_properties[prop+'_clahe'] = propperty[0,:]
```

```python
f, axs = plt.subplots(2,3,figsize=(16,8))
axs[0][0].imshow(img_ok)
axs[0][0].axis('off')
axs[0][1].imshow(gray_ok, cmap='gray')
axs[0][1].axis('off')
axs[0][2].imshow(clahe_ok, cmap='gray')
axs[0][2].axis('off')
#axs[0].title.set_text('(a)')
axs[1][0].imshow(img_defect)
axs[1][0].axis('off')
axs[1][1].imshow(gray_defect, cmap='gray')
axs[1][1].axis('off')
axs[1][2].imshow(clahe_defect, cmap='gray')
axs[1][2].axis('off')
#axs[1].title.set_text('(b)')
f.tight_layout()
plt.savefig('gclm_data.png', dpi=200)
plt.show()
```

```python
f, axs = plt.subplots(2,3,figsize=(14,7), sharey='row')
for axs_id, prop in zip(range(6),props):
    ax = axs[int(axs_id/3)][axs_id%3]
    #ax.title.set_text(prop)
    ax.set_xlabel(prop, fontsize=15)
    
    ax.plot(ok_properties[prop], degrees, 'r--', label='Good (Original)')
    ax.plot(defect_properties[prop], degrees, 'g--', label='Defect (Original)')
    
    ax.plot(ok_properties[prop+'_clahe'], degrees, 'r', label='Good (CLAHE)')
    ax.plot(defect_properties[prop+'_clahe'], degrees, 'g', label='Defect (CLAHE)')
    
    ax.yaxis.set_ticks(degrees)
    
    data = np.concatenate((ok_properties[prop], defect_properties[prop], ok_properties[prop+'_clahe'], defect_properties[prop+'_clahe']), axis=None)
    dist = (max(data)-min(data))/5
    labels = np.append(np.arange(min(data), max(data), dist), max(data))
    ax.xaxis.set_ticks(labels)
    if dist>1:
        ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%0.1f'))
    elif dist>0.01:
        ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%0.2f'))
    elif dist>0.001:
        ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%0.3f'))
    else:
        ax.xaxis.set_major_formatter(ticker.FormatStrFormatter('%.1e'))

axs[0][0].set_ylabel('Degrees')
axs[1][0].set_ylabel('Degrees')
ax.set_xlabel(prop)
axs[1][2].legend(loc='center right', framealpha=0.9)
f.tight_layout()
plt.savefig('gclm_properties.png', dpi=200)
plt.show()
```
